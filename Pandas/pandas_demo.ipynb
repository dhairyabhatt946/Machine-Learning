# Introduction to Pandas

## What is Pandas?

### A powerful library for data manipulation and analysis in Python. Built on top of Numpy, provides data structures like Series and DataFrame

## Why use Pandas?

### Easy data cleaning and transformation. Efficient handling of large datasets. Integration with other Python libraries like Matplotlib and Seaborn for visualization.

## Key Data Structures in Pandas

* Series (1 Dimensional)
* DataFrame (2 Dimensional)
